> 
> *Exercises*:
> 
> Begin by importing the right packages. Start by importing `matplotlib.pyplot` (for plotting), `numpy` (for binning and other stuff), 
> `random` (for generating random numbers), and `networkx` (for generating networks.)
> 
> * Binning real numbers
>   * Let's do a gentle start and use the `random` library generate 5000 data points from a Gaussian distribution with $\mu = 2$ and $\sigma = 0.125$.
>   * Now, let's use `numpy.histogram` to bin those number into 10 bins. What does the `numpy.histogram` function return? Do the two arrays have the same length?
>   * Then we use `matplotlib.pyplot.plot` to plot the binned data. You will have to deal with the fact that the counts- and bin-arrays have different lengths. Explain how you deal with this problem and why.
> * Binning integers
>   * But binning real numbers into a fixed number of bins is easy when `numpy.histogram` does all the work and finds the right bin boundaries for you. 
> Now we'll generate a bunch of integers and set the bin boundaries manually. This time, let's grab data from a Poisson distribution. As it turns out 
> `numpy` also has some convenient random number generators. Use `numpy.random.poisson` to generate 5000 numbers drawn from a Poisson distribution 
> characterized by $\lambda = 10$. Find the maximum and minimum value of your 5000 random numbers. 
>   * Instead of simply specifying the number of bins for `numpy.histogram`, let's specify the bins we want using a vector. 
> Create a vector $v$
> that results in a binning that puts each integer value in its own bin and where the first bin contains the minimum number you found above, 
> and the last bin contains the maximum number. 
> Use the vector by setting `numpy.histogram`'s `bin` parameter as `bin = ` $v$. 
> What is the sum over bin counts? Explain how the binning-vectors first and last element 
> relates to the min and max from the Poisson distribution.
>   * Now, use a bar chart (`matplotlib.pyplot.bar`) to plot the distribution 
> * Binning and plotting degree distributions. 
>   * Let's generate the Erdös-Renyi (ER) network which has a degree distribution that matches the Poisson distribution above. 
> First we have to figure out which values the ER parameters (_N_ and _p_) should assume. It's easy to see that $N = 5000$, but how do you find $p$? 
> **Hint**: The parameter $\lambda$ in the Poisson distribution corresponds to the average degree, so you have to find a $p$ that results in an average 
> degree, $k = 10$. And you know that $\langle k \rangle = p (N-1)$, which will give you $p$.
>   * Now, use `networkx` to create the graph and extract the degree distribution.
>   * Finally, create a nice bar plot of the degree distribution, including axes labels and a plot title. Make sure that it looks like the Poisson distribution you plotted above.
> 

```python
# Step 1: Import necessary packages
import matplotlib.pyplot as plt
import numpy as np
import random
import networkx as nx

# Step 2: Generate 5000 data points from a Gaussian distribution with mean 2 and std dev 0.125
mu = 2
sigma = 0.125
data_gaussian = np.random.normal(mu, sigma, 5000)

# Step 3: Bin the Gaussian data into 10 bins using numpy.histogram
counts, bin_edges = np.histogram(data_gaussian, bins=10)

# Step 4: Plot the binned data using matplotlib
# Dealing with different lengths of counts and bin_edges: bin_edges has 1 more element than counts, so we take the midpoints of bin_edges for plotting.
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Plot the binned data
plt.figure(figsize=(8, 6))
plt.plot(bin_centers, counts, marker='o')
plt.title("Binned Data from Gaussian Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 5: Generate 5000 numbers from a Poisson distribution with lambda=10
lambda_poisson = 10
data_poisson = np.random.poisson(lambda_poisson, 5000)

# Step 6: Find min and max values of Poisson data
min_val_poisson = np.min(data_poisson)
max_val_poisson = np.max(data_poisson)

# Step 7: Create a bin vector that puts each integer value in its own bin
v = np.arange(min_val_poisson, max_val_poisson + 2)  # Adding 1 to max, and another 1 to include the right edge

# Step 8: Bin the Poisson data using the custom bin vector
poisson_counts, poisson_bin_edges = np.histogram(data_poisson, bins=v)

# Step 9: Plot the Poisson distribution as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(v[:-1], poisson_counts, width=0.8, align='center')
plt.title("Poisson Distribution (lambda=10)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 10: Generate the ER network with N=5000 and p such that the average degree k=10
N = 5000
k = 10
p = k / (N - 1)

# Create the ER network
ER_network = nx.gnp_random_graph(N, p)

# Step 11: Extract the degree distribution
degrees = [deg for node, deg in ER_network.degree()]
degree_counts, degree_bin_edges = np.histogram(degrees, bins=v)

# Step 12: Plot the degree distribution as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(v[:-1], degree_counts, width=0.8, align='center')
plt.title("Degree Distribution of ER Network (N=5000, <k>=10)")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

```python
import networkx as nx
import random

# We want an average degree of 1.5. k = p(N-1). solve the inverse.

p = 1.5/199

G = nx.gnp_random_graph(n=200,p=p)

# Get the GCC
GCC = max(nx.connected_components(G),key=len)
GCC_copy=G.subgraph(GCC).copy()

# random node from it
start_node = random.choice(list(GCC_copy.nodes))

# find all nodes 2 steps away from the starting 
path_lengths = nx.single_source_shortest_path_length(GCC_copy, start_node)
nodes_two_steps_away = [node for node, length in path_lengths.items() if length == 2]
```

```python
import matplotlib.pyplot as plt
# Plot the GCC with specified colors
node_colors = []
for node in GCC_copy.nodes:
    if node == start_node:
        node_colors.append("#000000")  # Black for starting node
    elif node in nodes_two_steps_away:
        node_colors.append("#ff0000")  # Red for nodes 2 steps away
    else:
        node_colors.append("#A0CBE2")  # Blue for all other nodes

# Draw the graph
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(GCC_copy)  # Layout for a better visual representation
nx.draw(GCC_copy, pos, node_color=node_colors, with_labels=False, node_size=50)
plt.show()
```
