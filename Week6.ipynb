```python
# 4
# This code iterates over each file in the state_union corpus and counts occurrences of the target words
from nltk.corpus import state_union

words_to_count = ['men', 'women', 'people']

for fileid in state_union.fileids():
    speech = state_union.words(fileid)
    men_count = speech.count('men')
    women_count = speech.count('women')
    people_count = speech.count('people')
    print(f'{fileid}: men={men_count}, women={women_count}, people={people_count}')
```

```python
# 11
# This code calculates the frequency of modal verbs across different genres in the Brown corpus,
from nltk.corpus import brown
from nltk import FreqDist

genres = brown.categories()
modals = ['can', 'could', 'may', 'might', 'must', 'will']

for genre in genres:
    genre_words = brown.words(categories=genre)
    fdist = FreqDist(w.lower() for w in genre_words)
    print(f"Modal counts in {genre}:")
    for modal in modals:
        print(f'{modal}: {fdist[modal]}')
    print()
```

```python
# 15
# This code creates a frequency distribution of words in the Brown corpus and filters words that occur three or more times.

fdist = FreqDist(brown.words())

# Find words that occur at least three times
words_at_least_three = [word for word, count in fdist.items() if count >= 3]
print(words_at_least_three)
```

```python
# 17

# This function filters out stopwords from a text and returns the 50 most frequent non-stopwords.


from nltk.corpus import stopwords

def most_frequent_non_stopwords(text, n=50):
    stop_words = set(stopwords.words('english'))
    filtered_words = [w for w in text if w.lower() not in stop_words]
    fdist = FreqDist(filtered_words)
    return fdist.most_common(n)

text = brown.words(categories='news')  # example with 'news' genre
print(most_frequent_non_stopwords(text))
```

```python
# 18
# This code computes the most frequent bigrams in a text, ignoring bigrams that contain stopwords.
def most_frequent_non_stopword_bigrams(text, n=50):
    stop_words = set(stopwords.words('english'))
    filtered_bigrams = [(w1, w2) for w1, w2 in bigrams(text) 
                        if w1.lower() not in stop_words and w2.lower() not in stop_words]
    fdist = FreqDist(filtered_bigrams)
    return fdist.most_common(n)

text = brown.words(categories='news')  # example with 'news' genre
print(most_frequent_non_stopword_bigrams(text))
```

```python
# 23

import nltk
from nltk.corpus import brown
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

# Load a large text (e.g., the Brown corpus)
words = brown.words()

# Calculate word frequencies
freq_dist = Counter(words)

# Sort words by frequency in descending order
sorted_freq = sorted(freq_dist.values(), reverse=True)

# Rank of words
ranks = list(range(1, len(sorted_freq) + 1))

# Plot word frequency vs rank
plt.figure(figsize=(10, 6))
plt.plot(ranks, sorted_freq)
plt.xscale('log')
plt.yscale('log')
plt.title('Word Frequency vs Rank (Zipf\'s Law)')
plt.xlabel('Rank (log scale)')
plt.ylabel('Frequency (log scale)')
plt.grid(True)
plt.show()
```
